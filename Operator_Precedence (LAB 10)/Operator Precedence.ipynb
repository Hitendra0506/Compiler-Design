### Grammar -

* E -> E + E
* E -> E - E
* E -> E * E
* E -> E / E
* E -> ( E )
* E -> ~ E 
* E -> id

### Creating the precedence table

```python
import pandas as pd

precedence = dict()

precedence['id'] = 9
precedence['('] = 8
precedence['~'] = 7
precedence['/'] = 6
precedence['*'] = 5
precedence['+'] = 4
precedence['-'] = 3
precedence[')'] = 2
precedence['$'] = 1

ops = list(precedence.keys())
table = dict()
for i in ops:
    table[i] = dict()
    for j in ops:
        a = precedence[i]
        b = precedence[j]
        if a >= b:
            table[i][j] = '>'
        else:
            table[i][j] = '<'

table['-']['+'] = '>'
table['*']['/'] = '>'
table['~']['~'] = '<'
table['id']['id'] = '-'
table['id']['('] = '-'
for i in ops :
    table['('][i] = '<'
for i in ops :
    table[')'][i] = '>'
table['('][')'] = '='
table[')']['('] = '-'
table[')']['id'] = '-'
table['$'][')'] = '-'
table['$']['$'] = '-'
table['(']['$'] = '-'
df = pd.DataFrame(table).transpose()
df
```

### Reduction Procedure

```python
def reduce(s):
    if s in ['+','-','*','/']:
        return 'E -> E ' + s + ' E'
    if s in [')','(']:
        return 'E -> ( E )'
    if s == '~':
        return 'E -> ~ E'
    if s == 'id':
        return 'E -> id'
```

### Parsing Algorithm

```python
expression = '~ ( id + id ) * id'
expression = expression.split(' ')
expression.append('$')
ptr = 0
stack = ['$']
while True:
    top = stack[-1]
    point = expression[ptr]
    if top==point=='$':
        break
    if table[top][point] in ['<','=']:
        stack.append(point)
        ptr = ptr+1
    elif table[top][point] == '>':
        s = stack.pop()
        if s == ')':
            s = stack.pop()
        print('Reduce : ' + reduce(s))
    else:
        print('ERROR - Invalid Expression')
        break
```
